<h3>Transform some values to Logarithmic Scale</h3>
<p>Log. formula: <code>log(1+<b>x</b>)</code>.</p>

```python
train_set['SalePrice_log'] = None
train_set['SalePrice_log'] = np.log(1 + train_set['SalePrice'])
```

<h3>Cross-correlation</h3><p>Across all the numerical variables.<br><br>The high correlation coefficients (either <b>â‰¥ 0.5 or â‰¤ -0.4</b>) are marked out as follows. This gives us some useful insight to apply <b>PCA (principal component analysis)</b> on those highly correlated pairs/groups of numeric variables, to eliminate multi-colinearity among them, and hopefully may lead to better model prediction performance in the end. This would be one of promising future directions to work on.</p>

```python
train_set_numerical = train_set[['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual', 'OverallCond', 'YearBuilt',
                                'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF',
                                '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath',
                                'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
                                'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch',
                                '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold', 'SalePrice']].dropna()

threshold = 0.4

corr = train_set_numerical.corr()

sns.set()
f, ax = plt.subplots(figsize = (16, 14))
sns.heatmap(corr, annot = False, linewidths = 0.01, linecolor = '#cccccc',
            ax = ax, cmap = 'Greys', square = True, vmin = threshold)
```

<p>Some features are of very strong correlations with the <b>house sale price</b>, e.g. <b>Overall Quality</b>, <b>Exterior Quality</b>, <b>Basement Quality</b>, etc. while some others may be of very weak/low correlations such as <b>Lot Shape</b>, <b>Fence quality</b>, and <b>Overall Condition</b>, etc.</p>

```python
# Print Top Features from less significance to the highest correlated
top_features_all = pd.DataFrame(corr['SalePrice'])
top_features_all = top_features_all.sort_values('SalePrice').tail(len(corr)).drop(corr.index[len(corr)-1])
#print(top_features_all, type(top_features_all))
```

```python
_ = plt.barh(range(len(top_features_all)), top_features_all['SalePrice'].values, color = 'blue', alpha = 0.56)
_ = plt.yticks(range(len(top_features_all)), top_features_all.index)
_ = plt.axvline(x = 0, color = 'black')
_ = plt.axvline(x = threshold, color = 'black', linestyle = '--')

plt.show()
```

<p>We will pick the top <b>7</b> highest correlated ordinal features and converted them into numerical variables and dropped some of the lowest correlated variables, both of which can help improve the model prediction performance.</p>

<p>For nominal categorical variables, the box plots below show the house price range variations across different categorical levels/values.  We can see that for the two example categorical variables, i.e. <b>neighborhood</b> and <b>exterior type</b>, <b>house sale prices</b> may vary a lot for different levels/values. This result also implies that these features may have strong impact on the house sale prices.</p>

```python
ax = sns.boxplot(x = 'Neighborhood', y = 'SalePrice', data = train_set, saturation = 1, fliersize = 1.25)
plt.xticks(rotation=90)
plt.grid(which = 'major', alpha = 0.5)
plt.title('variable = Neighborhood')
plt.show()
```

```python
ax = sns.boxplot(x = 'Exterior1st', y = 'SalePrice', data = train_set, saturation = 1, fliersize = 1.25)
plt.xticks(rotation=90)
plt.grid(which = 'major', alpha = 0.5)
plt.title('variable = Exterior1st')
plt.show()
```

<h2>Feature Engineering</h2>

<p>Based on the data analysis results/findings, along with some good intuition or common knowledge on real estates, we conducted our first set of feature engineering steps as follows:
<ul>| Identified 7 ordinal variables with very high correlation with house sale price, and converted them into numerical variables.</ul>
<ul>| Added a "total square-feet" variable as the summation of total basement square-feet, and 1st and 2nd floor square feet.</ul>
<ul>| Dropped some zero-variance or near-zero-variance categorical variables (i.e. almost all of the same categorical value/level).</ul>
<ul>| Handled time related variables properly, e.g. garage year built, year built, year sold, month sold, to better fit for regression.</ul></p>
